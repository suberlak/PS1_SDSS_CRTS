Well.. I don't have SDSS S82 stellar light curves,  but I do have the CRTS light curves for the 10% of the S82 SDSS standard stars catalog ... 

```python
dir_in = '../../CRTS_PROJECT/raw_LC_CRTS/stars/'
crts_stars = os.listdir(dir_in)
crts_stars.remove('file.list')
crts_stars.remove('radec.00')
```

Consider only non-empty files ...

```python
# this takes 30 sec to run ....
to_remove = []
for star in crts_stars : 
    num_lines = sum(1 for line in open(dir_in + star, 'r'))
    if num_lines < 10 :
        to_remove.append(star) 
print('We have %d  files with less than 10 lines ' %len(to_remove))   
```

```python
for x in to_remove : 
    crts_stars.remove(x)
```

```python
print('We are left with ', len(crts_stars), ' non-empty light curves')
```

```python
def findMAP(lc):
    prior = 'Jeff1' # should have used Jeff1 , anyways ! 
    SF_inf = 0.2  # mags 
    sigma_in = SF_inf / np.sqrt(2)
    tau_in = 575 # days 
    sig_lims , tau_lims = [0.01,1.0], [1,5000]
    
    # find the MAP estimate 
    sigmaMAP, tauMAP,gp = mod.find_celerite_MAP(lc['mjd'],lc['mag'],lc['magerr'], 
                              sigma_in, tau_in, prior=prior,set_bounds=True, 
                              sig_lims=sig_lims, tau_lims=tau_lims ,verbose=False)

    return sigmaMAP, tauMAP

```

```python
s,t = [],[]
for i,star in enumerate(crts_stars) :
    if i % 100 == 0:
        print(i)
    lc = Table.read(dir_in+star, format='ascii', names=['mjd', 'mag', 'magerr', '0'])
    sigmaMAP, tauMAP = findMAP(lc)
    s.append(sigmaMAP)
    t.append(tauMAP)
```
